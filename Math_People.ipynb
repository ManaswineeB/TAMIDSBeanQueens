```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
people = requests.get('https://api.library.tamu.edu/scholars-discovery/individual/n73bc0d79?fbclid=IwAR14AfAGbKtQYKRKPsaNcAdzvqQFt76bXYunrdSUMBxZXpGKC6yKxkcGQ4E').json()
```

```python
people = people['people']
```

```python
publication_data=[]
for i in people:
    url = 'https://api.library.tamu.edu/scholars-discovery/individual/'+i['id']
    publication_data.append(requests.get(url).json())
```

```python
keywords = set()
for i in publication_data:
    if 'publications' in i.keys():
        for j in i['publications']:
            url = 'https://api.library.tamu.edu/scholars-discovery/individual/'+j['id']
            publication = requests.get(url).json()
            if 'keywords' in publication.keys():
                for k in publication['keywords']:
                    keywords.add(k)
            else:
                continue
    else:
        continue
```
